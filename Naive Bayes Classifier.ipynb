```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import CategoricalNB
from sklearn.metrics import confusion_matrix
```

```python
df=pd.read_csv('C:/Users/Noor/Desktop/college/term 4/Pattern recognition/adult/adult test.csv')
df.head(10)
```

```python
# Split features and target variable
features = df.drop('income', axis=1)
target = df['income']
```

```python
# Drop rows with missing values
df=df.dropna()

# Convert categorical variables into dummy/indicator variables
dummies=pd.get_dummies(features, drop_first=True)
dummies.head(3)
```

```python
target=pd.get_dummies(target,drop_first=True)
target.head(5)
```

```python
features=features.drop(['workclass','education','marital_status','occupation','relationship','race','sex','country'],axis='columns')
features.head(3)
```

```python
features=pd.concat([features,dummies],axis='columns')
features.head(5)
```

```python
# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.2, random_state=80)
```

```python
#Train Categorical Naive Bayes Classifier
categorical_naive_bayes = CategoricalNB()
categorical_naive_bayes.fit(X_train, y_train)
```

```python
categorical_naive_bayes.score(X_test,y_test)
```

```python
# Predictions on the test set
y_pred = categorical_naive_bayes.predict(X_test)
```

```python
# Compute confusion matrix
conf_matrix = confusion_matrix(y_test, y_pred)
TN, FP, FN, TP = conf_matrix.ravel()
```

```python
print(conf_matrix)
```

```python
# Compute Sensitivity and Specificity
sensitivity = TP / (TP + FN)
specificity = TN / (TN + FP)

print("Sensitivity (True Positive Rate):", sensitivity)
print("Specificity (True Negative Rate):", specificity)
```

```python
# Calculate posterior probabilities for each instance
posterior_probabilities = categorical_naive_bayes.predict_proba(X_test)[:, 1]
```

```python
# Example of computing posterior probability for a single instance (adjust as needed)
instance_index = 0
posterior_prob = posterior_probabilities[instance_index]

print("Posterior Probability of making over 50K:", posterior_prob)
print("Posterior Probability of making less than 50K:",(1-posterior_prob))
```
